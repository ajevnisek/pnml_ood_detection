```python
import os.path as osp

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
# Plot params
# mpl.style.use("seaborn-dark")
mpl.style.use("ggplot")
mpl.rcParams["axes.grid"] = True
mpl.rcParams["legend.frameon"] = True
mpl.rcParams["figure.dpi"] = 85
mpl.rcParams["axes.labelsize"] = "large"
# plt.rcParams["ytick.major.size"] = 5.0
# plt.rcParams["ytick.minor.size"] = 3.0
plt.rcParams["legend.framealpha"] = 0.9
w, h = plt.rcParams.get('figure.figsize')
```

```python
model_name = "DenseNet"
paths = [
    "../outputs/baseline_densenet_cifar100_20210531_104742/singular_values.pt",
    "../outputs/baseline_densenet_cifar10_20210531_105106/singular_values.pt",
    "../outputs/baseline_densenet_svhn_20210531_105327/singular_values.pt",
]

model_name = "ResNet"
paths = [
    "../outputs/baseline_resnet_cifar100_20210531_105824/singular_values.pt",
    "../outputs/baseline_resnet_cifar10_20210531_105718/singular_values.pt",
    "../outputs/baseline_resnet_svhn_20210531_105548/singular_values.pt",
]


s_list = [torch.load(path, map_location=torch.device("cpu")).numpy() for path in paths]
```

```python
plt.plot(s_list[0], label="CIFAR-100", alpha=0.75)
plt.plot(s_list[1], label="CIFAR-10", alpha=0.75)
plt.plot(s_list[2], label="SVHN", alpha=0.75)
plt.yscale("log")
plt.xlabel("Eigenvalue index")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.savefig(f"../outputs/visualizations/{model_name}_svd.pdf")
plt.show()
```
